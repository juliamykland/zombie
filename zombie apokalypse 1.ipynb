```python
from pylab import *
   
#Tidsstegs 
t0 = 0
N = 100000
tid = 400

dt = tid/(N-1)
b = 1500      #Bæreevne


m0 = 500
z0 = 10
dr = 1/(12*70)      #dødsraten
fr = dr*2          #fødselsraten, tidskalaen er målt i måneder, og man regner med menneskene lever i 70 år
mz =  0.005/(N/tid)      #menneske møter zombie
innv = 20/(N/tid)   #Innvandring
k = fr - dr        #reproduskjonsrate



#Matriser
t = zeros(N)
m = zeros(N)
mder = zeros(N)
z = zeros(N)


t[0] = t0
m[0] = m0
z[0] = z0



for i in range(N-1):
    mder[i] = k*m[i]
    m[i+1] = m[i] + (mder[i] * dt + innv) * (1 - m[i]/b) - mz * m[i] * z[i]
    z[i+1] = z[i]
    t[i+1] = t[i] + dt
    
    
#Plotting
fig = figure()
ax = fig.add_subplot(111)
data1 = ax.plot(t, m, '-b', label = 'Mennesker')
ax2 = ax.twinx()
data2 = ax2.plot(t, z, '-r', label = 'Zombier')

data = data1 + data2
datatittel = [l.get_label() for l in data]
ax.legend(data, datatittel, loc = 0)

ax.grid()
ax.set_xlabel('Tid (månder)')
ax.set_ylabel("Antall mennesker")
ax2.set_ylabel("Antall zombier")
ax2.set_ylim(0,40)
ax.set_ylim(0, b + 100)
plt.show()



```

```python
from pylab import *
   
#Tidsstegs 
t0 = 0
N = 100000
tid = 400

dt = tid/(N-1)
b = 1500      #Bæreevne


m0 = 500
z0 = 10
dr = 1/(12*70)      #dødsraten
fr = dr*2          #fødselsraten, tidskalaen er målt i måneder, og man regner med menneskene lever i 70 år
mz =  0.0005/(N/tid)      #menneske møter zombie
innv = 20/(N/tid)   #Innvandring
k = fr - dr        #reproduskjonsrate
bitt = 0.1      #andelen av de som møter zombie som blir til zombier
zd = 0.0009/(N/tid)           #andel zombier drept


#Matriser
t = zeros(N)
m = zeros(N)
mder = zeros(N)
z = zeros(N)
zder = zeros(N)


t[0] = t0
m[0] = m0
z[0] = z0



for i in range(N-1):
    mder[i] = k*m[i]
    m[i+1] = m[i] + (mder[i] * dt + innv) * (1 - m[i]/b) - mz * m[i] * z[i]
    z[i+1] = z[i] + zder[i] * dt + (mz * m[i] * z[i]) * bitt  - (zd  * z[i])
    t[i+1] = t[i] + dt
    

    
#Plotting
fig = figure()
ax = fig.add_subplot(111)
data1 = ax.plot(t, m, '-b', label = 'Mennesker')
ax2 = ax.twinx()
data2 = ax2.plot(t, z, '-r', label = 'Zombier')

data = data1 + data2
datatittel = [l.get_label() for l in data]
ax.legend(data, datatittel, loc = 0)

ax.grid()
ax.set_xlabel('Tid (månder)')
ax.set_ylabel("Antall mennesker")
ax2.set_ylabel("Antall zombier")
ax2.set_ylim(0,100)
ax.set_ylim(0, b + 100)
plt.show()
```

```python
from pylab import *
   
#Tidsstegs 
t0 = 0
N = 100000
tid = 400

dt = tid/(N-1)
b = 1500      #Bæreevne


m0 = 500
z0 = 10
dr = 1/(12*70)      #dødsraten
fr = dr*2          #fødselsraten, tidskalaen er målt i måneder, og man regner med menneskene lever i 70 år
mz =  0.0005/(N/tid)      #menneske møter zombie
innv = 20/(N/tid)   #Innvandring
k = fr - dr        #reproduskjonsrate
bitt = 0.1      #andelen av de som møter zombie som blir til zombier
zd = 0.0009/(N/tid)           #andel zombier drept


#Matriser
t = zeros(N)
m = zeros(N)
mder = zeros(N)
z = zeros(N)
zder = zeros(N)


t[0] = t0
m[0] = m0
z[0] = z0



for i in range(N-1):
    if m[i] < 200:
        m[i] = m[i] + 100
        zd = 2 * zd
    mder[i] = k*m[i]
    m[i+1] = m[i] + (mder[i] * dt + innv) * (1 - m[i]/b) - mz * m[i] * z[i]
    z[i+1] = z[i] + zder[i] * dt + (mz * m[i] * z[i]) * bitt  - (zd * z[i])
    t[i+1] = t[i] + dt
    

    
#Plotting
fig = figure()
ax = fig.add_subplot(111)
data1 = ax.plot(t, m, '-b', label = 'Mennesker')
ax2 = ax.twinx()
data2 = ax2.plot(t, z, '-r', label = 'Zombier')

data = data1 + data2
datatittel = [l.get_label() for l in data]
ax.legend(data, datatittel, loc = 0)

ax.grid()
ax.set_xlabel('Tid (månder)')
ax.set_ylabel("Antall mennesker")
ax2.set_ylabel("Antall zombier")
ax2.set_ylim(0,100)
ax.set_ylim(0, b + 100)
plt.show()
```
