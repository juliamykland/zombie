# Zombie apokalypse

```python
from pylab import *
    
m = 500                     #Mennesker
dr = 1/(12*70)      #dødsraten
fr = dr*5     #fødselsraten, tidskalaen er målt i måneder, og man regner med menneskene lever i 70 år
innv = 3                   #innvandring i måneden
maxm = 5000

#lister
antall_mnd = []
populasjon_liste = []

def population(n,m):        #n står for number of months
    global fr, dr, innv
    m0 = m
    mr = m
    for i in range(n):
        m0 += m0*(fr-dr) + innv
        mr  = mr + (mr * fr - mr * dr + innv) * (1-(mr/maxm))    #ressursbegrensning
    return  m0, mr

print(population(100,m))

for i in range(1000):
    antall_mnd.append(i)
    populasjon_liste.append(population(i,m))
    
plot(antall_mnd, populasjon_liste)

title("")
xlabel("antall månder")
ylabel("populasjon")
ylim(0, 6000)
show()
```
